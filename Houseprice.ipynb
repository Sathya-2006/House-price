```python
# data preprocessing
list1=['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea']
```

```python
df[list1]=df[list1].replace({'yes':1,'no':0})
```

```python
df['furnishingstatus']=df['furnishingstatus'].replace({'furnished':2,'semi-furnished':1,'unfurnished':0})
```

```python
# standiardization
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
list2=['price','area']
```

```python
df[list2]=scaler.fit_transform(df[list2])
```

```python
x=df.drop('price',axis=1)

y=df['price']
```

```python
x.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
lr_accuracy=r2_score(y_test,y_pred)*100
```

```python
lr_accuracy
```

```python
# lets improve your model using hyperparameter tuning!
# Linear Regression has no major hyperparameters to tune

from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
```

```python
rndfrt=RandomForestRegressor()
```

```python
param_grid={
    'n_estimators':[50,100,200],
    'max_depth':[None,5,10,20],
    'min_samples_split':[2,5,10],
    'min_samples_leaf':[1,2,4]

}
```

```python
grid_search=GridSearchCV(
    estimator=rndfrt,
    param_grid=param_grid,
    cv=5,
    scoring='r2',
    n_jobs=-1,
    verbose=2
)
```

```python
grid_search.fit(x_train,y_train)
```

```python
print("Best Parameters:", grid_search.best_params_)
```

```python
best_model = grid_search.best_estimator_
y_pred = best_model.predict(x_test)

from sklearn.metrics import r2_score
print("Improved Accuracy:", r2_score(y_test, y_pred)*100)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV
```

```python
model = GradientBoostingRegressor()
```

```python
param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [2, 3, 4],
    'subsample': [0.8, 1.0],
}
```

```python
grid_search = GridSearchCV(
    estimator=model,
    param_grid=param_grid,
    scoring='r2',
    cv=5,
    n_jobs=-1,
    verbose=2
)
```

```python
grid_search.fit(x_train, y_train)
```

```python
print("Best Params:", grid_search.best_params_)
```

```python
best_model = grid_search.best_estimator_
```

```python
y_pred = best_model.predict(x_test)

print("Improved Accuracy:", r2_score(y_test, y_pred)*100)
```
